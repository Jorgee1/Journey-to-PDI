# Punto 2

```python

# Funcin para rotar imagen ##################################################################
# el parametro flag controla la rotacin
# 1 - Horizontal y 0 - Vertical
def flipar(img,flag):
    if(flag==0):                # Se compara el flag para saber que funcin aplicar 
        return np.fliplr(img)   # Se regresa la imagen rotada verticalmente
    else:
        return np.flipud(img)   # Se regresa la imagen rotada horizontalmente
##############################################################################################

img = cv2.imread('Woman.bmp',cv2.IMREAD_GRAYSCALE)   # Se carga la imagen en la variable "img"

print("Imagen original")
plt.imshow(img,cmap="gray")
plt.show()                                           # Se muestra la imagen original                                           

print("Imagen rotada horizontalmente")
rotadaH = flipar(img,1)                              # Se aplica la rotacion horizontal a la imagen   
plt.imshow(rotadaH, cmap="gray")
plt.show()                                           # Se muestra la imagen alterada 

print("Imagen rotada verticalmente")
rotadaV = flipar(img,0)                              # Se aplica la rotacion vertical a la imagen  
plt.imshow(rotadaV, cmap="gray")
plt.show()                                           # Se muestra la imagen alterada 
```

# Punto 3

```python

# Funcin para generar el negativo de una imagen #########################################################################
def negativo(pic):
    height, width = pic.shape                             # Se guarda las dimensines de la imagne
    for row in range(height):                             # Se recorren todos los pixeles
        for column in range(width):
            pic[row][column] = 255 - pic[row][column]     # Se remplaza cada pixel por el inverso
    return pic                                            # Sr regresa la imagen alterada
##########################################################################################################################

img = cv2.imread('moon.jpg',cv2.IMREAD_GRAYSCALE)   # Se carga la imagen en la variable "img"
plt.imshow(img,cmap="gray")

print("Imagen original")
plt.show()                                          # Se muestra la imagen original

print("Negativo de la imagen")
plt.imshow(negativo(img), cmap="gray")              # Se llama la funcoin "negativo"
plt.show()                                          # Se muestra la imagen alterada
```

# Punto 4

```python

# Funcin para obtener la intensidad promedio de la imagen #####################################################
def promedio(img):                           # la funcion toma una imagen como parametro
    
    acc = 0                                  # Acumulador que donde se guarda la suma de todos los pixeles
                                             # los pixeles
    
    height, width = img.shape                # Se guardan las dimensiones de la imagen
    for row in range(height):                # Se recorren todos los pixeles de la imagen con los ciclos for
        for column in range(width):
            acc = acc + img[row][column]
    return (acc/(height * width))            # Se regresa el suma de todos los pixeles sobre el numero total
################################################################################################################

print("Imagen 1")

img = cv2.imread('moon.jpg',cv2.IMREAD_GRAYSCALE)                       # Se carga la imagen de la luna
plt.imshow(img,cmap="gray")  
plt.show()                                                              # Se muestra la imagen

print("La intensidad promedio de la imagen es: " + str(promedio(img)))  # Se llama a la funcin promedio
print("\nImagen 2")
img = cv2.imread('origami.jpg',cv2.IMREAD_GRAYSCALE)                    # Se carga la imagen de pikachu
plt.imshow(img,cmap="gray")
plt.show()                                                              # Se muestra la imagen

print("La intensidad promedio de la imagen es: " + str(promedio(img)))  # Se llama a la funcin promedio
```

## Conclusin

El valor promedio fue establecido entre los limites de 0 a 255, donde cero es negro y 255 es blanco. 
entre mas bajo es el valor promedio la imagen es mas oscura (Imagen 1) y entre mas alto es el valor, esta es mas clara (Imagen 2).

# Punto 5

```python

# Funcin para obtener una versin de la imagen a blanco y negro ###################################################
# En la documentacin 0 y 255 son la convencin para delimitar los valores minimos y maximos 
# respectivoamente de una imagen de 8 bits. Donde 0 es negro y 255 es blanco.

def bin_img(img,lim):                     # Se toman dos paarametros, la imagen y un limite definido por el usuario      
    pic = img
    height, width = pic.shape             # Se guardan las dimensiones de la imagen
    for row in range(height):             # Se recorren todos los pixeles de la imagen
        for column in range(width):
            if lim < pic[row][column]:    # Se compara el valor del limite con el valor de los pixeles
                pic[row][column] = 255    # Blanco si supera el limite
            else:
                pic[row][column] = 0      # Negro si no supera el limite
            
    return pic                            # Se regresa la imagen alterada
####################################################################################################################
th = 150                                                                # Threshold de la imagen 
print("Imagen original")

img = cv2.imread('moon.jpg',cv2.IMREAD_GRAYSCALE)                       # Se carga la imagen
plt.imshow(img,cmap="gray")  
plt.show()                                                              # Se muestra la imagen

print("Imagen alterada")

plt.imshow(bin_img(img,th),cmap="gray")                                 # Se aplica la funcin en la imagen
plt.show()
```

# Punto 6 

```python

# Funcin para obtener una versin de la imagen a blanco y negro #########################################################
# utilizando el promedio como threshold

def ave_int(img):
    p = promedio(img)      # Se llama a la funcin "promedio" para calcular la intensidad promedio de la imagen 
    bi = bin_img(img,p)    # Se llama la funcin "bin_img" para obtener una versin de la imagen a blanco y negro
    return bi              # Se regresa la imagen alterada
#########################################################################################################################

print("Imagen original")

img = cv2.imread('niagara.jpg',cv2.IMREAD_GRAYSCALE)                       # Se carga la imagen
plt.imshow(img,cmap="gray")  
plt.show()                                                              # Se muestra la imagen

print("Imagen alterada")

img = cv2.imread('niagara.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(ave_int(img),cmap="gray")
plt.show()
```
