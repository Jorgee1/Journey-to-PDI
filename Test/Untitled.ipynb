```python
img = cv2.imread("image.jpg",cv2.IMREAD_GRAYSCALE)
plt.imshow(img,cmap="gray")
plt.show()

#Define flip_image function
def flip_image( Image ):   
    w,h =Image.shape
    temp = 0
    for i in range(0,w):
        for j in range(0,int(round(h/2))):
            temp = Image[i][j]
            Image[i][j] = Image[i,h-j-1]
            Image[i,h-j-1] = temp

    return Image

    
plt.imshow(flip_image(img),cmap="gray")
plt.show()
```
